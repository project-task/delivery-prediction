The idea is the following, long_text_prep() function is used for text preprocessing: 

   * Keep only european chars and whitespaces with no digits
   * remove single letter chars
   * convert all whitespaces (tabs etc.) to single wspace
    
Then we tokenize and run stemmer also getting rid of stopwords in the process. After this has been done, we can apply TF-IDF on the column and create matrix with calculated values that we can use for the further training of the model. Also embeddings and word2vec can be used for the similar purpose as well.
```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def long_text_prep(text):
    
    WS = re.compile(r"\s+", re.IGNORECASE)
    AS = re.compile(r"[^A-Za-zÀ-ž ]", re.IGNORECASE)
    SI = re.compile(r"\b[A-Za-zÀ-ž]\b", re.IGNORECASE)

    text = re.sub(WS, " ", text)
    text = re.sub(AS, " ", text)
    text = re.sub(SI, " ", text)

    tokens_word = word_tokenize(text)
    tokens_lower = [x.lower() for x in tokens_word]

    words = [stemmer.stem(x) for x in tokens_lower if x not in stop_words]

    text_prep = " ".join(words)
    return text_prep
```

For the preparation for model, sklearn's pipeline will be used. We will have two pipelines: first is for dropping certain features that we are not planning to use in the model and the second is the categorical preprocessing and then feature union of the feature set.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.base import BaseEstimator, TransformerMixin
```

We need to have couple of estimators. The idea is to use sklearn base estimators and create our own estimators using classes. In the production code, they should be separated in different python modules and link with each other for main preprocessing module in our code.

```python
class DropFeature:

    def __init__(self, variables_to_drop=None):
        self.variables = variables_to_drop

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        '''
        Dropping the selected columns
        '''
        X = X.copy()
        X = X.drop(self.variables, axis=1)

        return X
```

```python
to_drop = ['order_id', 'customer_id', 'created_timestamp_local', 'has_instruction', 
           'order_value', 'order_items_count']

drop_pipeline=Pipeline([
    
('drop',DropFeature(to_drop))

                      ])

X_train = drop_pipeline.transform(X_train)
X_test = drop_pipeline.transform(X_test)
```

We need to have the estimator that can help to select the certain columns. As well, for highly cardinality categorical features is always good idea trying to use LeaveOneOut Encoder, because it can calculate the shares of values and create quite effective distribution. We will use category_encoders realisation, therefore we will need an estimator for that:

```python
#Intended to select the certain columns from dataframe
class Selector(BaseEstimator, TransformerMixin):
    
    def __init__(self, key):
        self.key = key
        
    def fit(self, x, y=None):
        return self
    
    def transform(self, df_d):
        return df_d[self.key]
```

```python
from category_encoders import  LeaveOneOutEncoder
```

```python
#LeaveOneOutEncoder designed to be used as sklearn estimator
class LeaveOOEncoder(BaseEstimator, TransformerMixin):

    def fit(self, X, y, **fit_params):
        loo = LeaveOneOutEncoder()
        self.loo = loo.fit(X, y)

        return self
    
    def transform(self, X, **fit_params):

        loo_data = self.loo.transform(X)
        return loo_data

    def fit_transform(self, X,y=None, **fit_params):
        self.fit(X, y, **fit_params)
        return self.transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import FeatureUnion

from sklearn.preprocessing import OneHotEncoder
```

In the main preprocessing pipeline, we will apply LeaveOOEncoder() to high caridnality categories and OneHotEncoder() for rest categorical features.

```python
pipeline = Pipeline([

    ('union', FeatureUnion(
        transformer_list=[

            ('categorical_high_cardinality_', Pipeline([
            ('selector_cat_high_', Selector(key=['delivery_geohash_precision8', 'delivery_postal_code'])),
            ('leaveoo_', LeaveOOEncoder())

            ])),
            
            ('categorical_', Pipeline([
            ('selector_cat_', Selector(key=['has_phone_number', 'has_call_instruction',
                                           'has_leave_instruction', 'has_lift_instruction', 'has_lobby_instruction',
                                           'has_gate_instruction', 'has_knock_instruction', 'has_bell_instruction'])),
            ('one_hot_', OneHotEncoder())

            ]))
    
]))
])
```

```python
X_train = pipeline.fit_transform(X_train, y_train)
X_test = pipeline.transform(X_test)
```

```python
X_train = X_train.toarray()
X_test = X_test.toarray()
```

```python
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import Ridge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
```

So let's test 4 estimators: Ridge regression, BayesianRidge (which should waork fine, because bayes models perform well in the situation of many independent categorical variables) and 2 ensembles: Random Forest and Gradient Boosting.

```python
regressors_m = [
    BayesianRidge(),
    Ridge(),
    GradientBoostingRegressor(),
    RandomForestRegressor()
]

names = [re.match(r"[^\(]+", name.__str__())[0] for name in regressors_m]
print(f"Regressors to test: {names}")
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
%%time
res_rmse = {}
res_r2 = {}

for name, reg in zip(names, regressors_m):
    print(f"Training regressor: {name}")
    reg.fit(X_train, y_train)
    pred = reg.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test,pred))
    res_rmse[name] = rmse
    r2 = r2_score(y_test,pred)
    res_r2[name] = r2
```

```python
for x, z in res_rmse.items():
    print(f"Results RMSE for {x}:")
    print(f"{z}\n")
```

```python
for x, z in res_r2.items():
    print(f"Results R2 for {x}:")
    print(f"{z}\n")
```
